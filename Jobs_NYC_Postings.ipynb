### Data Acquisition and Inspection:

```python
import pandas as pd
catalog_ = pd.read_csv("Downloads/Jobs_NYC_Postings.csv")
catalog_
```

```python
null_values_count = catalog_.isnull().sum()
print(null_values_count)
```

### Data Cleaning and Preparation:

```python
catalog_.drop(columns=['Recruitment Contact', 'Post Until', 'Work Location 1', 'Hours/Shift'], inplace=True)
catalog_.head()
```

```python
null_values_count = catalog_.isnull().sum()
print(null_values_count)
```

```python
catalog_['Preferred Skills'].fillna('Not specified', inplace=True)
catalog_.head()
```

```python
catalog_['Full-Time/Part-Time indicator'].fillna('Not specified', inplace=True)
catalog_.head()
```

```python
catalog_['Additional Information'].fillna('Information not provided', inplace=True)
catalog_.head()
```

```python
catalog_['To Apply'].fillna('See job posting for application details', inplace=True)
catalog_.head()
```

```python
catalog_['Minimum Qual Requirements'].fillna('Information not provided', inplace=True)
catalog_.head()
```

```python
null_values_count = catalog_.isnull().sum()
print(null_values_count)
```

### Exploratory Data Analysis (EDA) Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Visualization 1: Salary Distribution by Civil Service Title

```python
top_20_titles = catalog_.groupby('Civil Service Title')['# Of Positions'].sum().nlargest(20).index
filtered_catalog = catalog_[catalog_['Civil Service Title'].isin(top_20_titles)]

plt.figure(figsize=(14, 10))
sns.violinplot(y='Civil Service Title', x='Salary Range From', data=filtered_catalog, cut=0)
plt.title('Salary Distribution by Top 20 Civil Service Titles')
plt.xlabel('Salary Range From')
plt.ylabel('Civil Service Title')
plt.show()
```

#### Visualization 2: Get the top 20 job categories based on the total number of positions

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=catalog_, x='Posting Type')
plt.title('Distribution of Posting Type')
plt.xlabel('Posting Type')
plt.ylabel('Count')
plt.show()
```

#### Visualization 3: Salary Range Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=catalog_, x='Salary Range From', bins=30, kde=True)
plt.title('Salary Range Distribution')
plt.xlabel('Salary Range From')
plt.ylabel('Frequency')
plt.show()
```

#### Visualization 4: Job Category Counts

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=catalog_, y='Job Category', order=catalog_['Job Category'].value_counts().index[:10])
plt.title('Top 10 Job Categories')
plt.xlabel('Count')
plt.ylabel('Job Category')
plt.show()
```

#### Visualization 5: Agency-wise Distribution of Jobs

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=catalog_, y='Agency', order=catalog_['Agency'].value_counts().index[:10])
plt.title('Top 10 Agencies with Job Postings')
plt.xlabel('Count')
plt.ylabel('Agency')
plt.show()
```

#### Visualization 6: Distribution of Full-Time/Part-Time Jobs

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=catalog_, x='Full-Time/Part-Time indicator')
plt.title('Full-Time/Part-Time Jobs Distribution')
plt.xlabel('Full-Time/Part-Time')
plt.ylabel('Count')
plt.show()
```

#### Visualization 7: Relationship between Salary Range and Number of Positions

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=catalog_, x='Salary Range From', y='# Of Positions')
plt.title('Relationship between Salary Range and Number of Positions')
plt.xlabel('Salary Range From')
plt.ylabel('Number of Positions')
plt.show()
```

#### Visualization 7: Pairplot of Numerical Features

```python
sns.pairplot(catalog_[['# Of Positions', 'Salary Range From', 'Salary Range To']])
plt.title('Pairplot of Numerical Features')
plt.show()
```

#### Visualization 8: Correlation Heatmap of Numerical Features:

```python
corr_matrix = catalog_[['# Of Positions', 'Salary Range From', 'Salary Range To']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
import pandas as pd
df = catalog_  

fields_for_recommendation = ['Business Title', 'Salary Range From', 'Salary Range To']

df = df[fields_for_recommendation].dropna()
def recommend_salary_range(business_title, df=df):
    filtered_df = df[df['Business Title'] == business_title]

    if filtered_df.empty:
        return None, None

    # Calculate median salary range from and to for the given business title
    median_salary_from = filtered_df['Salary Range From'].mean()
    median_salary_to = filtered_df['Salary Range To'].mean()

    return median_salary_from, median_salary_to

sample_business_title = input("Enter the Business Title: ")
salary_range_from, salary_range_to = recommend_salary_range(sample_business_title)

if salary_range_from is not None and salary_range_to is not None:
    print(f"For the business title '{sample_business_title}':")
    print(f"Recommended Salary Range From: {salary_range_from}")
    print(f"Recommended Salary Range To: {salary_range_to}")
else:
    print(f"No salary information found for the business title '{sample_business_title}'")
```
